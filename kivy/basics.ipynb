```python
from kivy.app import App
from kivy.uix.label import Label
from kivy.uix.gridlayout import GridLayout
from kivy.uix.button import Button

class example_grid_button(GridLayout):
    def __init__(self,**kwargs):
        super(example_grid,self).__init__(**kwargs)
        
        self.cols=2
        self.inside=GridLayout()
        self.cols=2
        
        self.inside.add_widget(Label(text="First Name: "))
        self.fname=TextInput(multiline=False)
        self.inside.add_widget(self.fname)
        
        self.inside.add_widget(Label(text="Last Name: "))
        self.lname=TextInput(multiline=False)
        self.iside.add_widget(self.lname)
        
        self.inside.add_widget(Label(text="Email "))
        self.email=TextInput(multiline=False)
        self.inside.add_widget(self.email)
        
        self.submit= Button(text="Submit", font_size=40)
        self.submit.bind(on_press=self.submit_pressed)
        self.add_widget(self.submit)
        
    def submit_pressed(self,instance):
        fname=self.fname.text
        lname=self.lname.text
        email=self.email.text
        print("First Name: ", fname )
        print("Last Name: ", lname )
        print("Email: ", email )
        self.fname.text=""
        self.lname.text=""
        self.email.text=""
        
        
        
class myapp(App):
    def built(self):
        return example_grid_button()
myapp().run()
```
