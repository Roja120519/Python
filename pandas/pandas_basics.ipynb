- three principal components
    - the data
    - rows 
    - columns
- Programs
    - creating a dataframe

```python
#Pandas DataFrame can be created from the lists.
import pandas as pd
def create_DF_list():
    l=[100,110,120,130,140]
    name=["roja","nani","lally","naren","prakash"]
    df=pd.DataFrame(l,name)
    return df[0]
create_DF_list()
```

```python
#Pandas DataFrame can be created from the dictionaries
import pandas as pd
def create_DF_list():
    data={"name":["nani","lally","naren","prakash","roja"],"marks":[90,76,95,89,90],"branch":["ECE","Mech","ECE","Civil","CSE"],"address":["Karnataka","Karnool","surathkal","bhimavaram","Nuzvid"]}
    
    df=pd.DataFrame(data)
    #print(df)
    return df
    
create_DF_list()
```

#### Rows and columns in dataframe
- Data frame is a two-dimensional data structure
- columns
    - Retrive selected columns data
    - Adding columns to data frame
    - column deletion
- rows
    - Pandas provide a unique method to retrieve rows from a Data frame. DataFrame.loc[] method is used to retrieve rows from Pandas DataFrame. 
    - Rows can also be selected by passing integer location to an iloc[] function
    - dataframe.loc[] -> Used to rtrieve particular row values
    - iloc[] -> Used to 
    - Adding row to the dataframe
    - retrieve selected rows data
    - row deletion 
 - selecting multiple columns and rows from dataframe by giving index values
 - modifying values by using iloc function.

```python
# Printing two columns from the data frame
data=create_DF_list()
print(data[['name','marks']])
```

```python
# Adding a column to dataframe
age=[22,18,22,18,19]
data['age']=age
data
```

```python
# retrieve particular column values
name=data['name']
name
```

```python
# column deletion
data.drop(["marks"], axis = 1, inplace = True)
data
```

```python
# Row deletion
print(data)
data.drop(1)
```

```python
# retrieve particular row values
nani=data.loc[2]
nani
```

```python
roja=data.iloc[4]
roja
```

```python
# Adding a row
print(data)
new_row=pd.DataFrame({'name':'likhi','branch':'EEE','address':'viayawada'},index=[0])
data=pd.concat([new_row,data]).reset_index(drop=True)
data
```

```python
# Adding a row
print(data)
new_row=pd.DataFrame({'name':'likhitha','branch':'EEE','address':'viayawada','age':19},index=[0])
data=pd.concat([new_row,data]).reset_index(drop=True)
data
```

```python
data.drop([1])
data.drop([2])
```

```python
# selecting multiple rows from dataframe by giving index values 
# df.iloc[[rows index values seperated by commas],[columns index vlaues seperated by commas]]

r=data.iloc[[1,5],[0,2]]
r
```

```python
#selecting multiple columns from dataframe by giving index values
c=data.iloc[[0,1,2,3],[3,1,2]]
c
```

```python
# Modifying values

data.iloc[[0],[0]]='EEE'

#data=data.drop([3])
#data=data.drop([4])
data
```
