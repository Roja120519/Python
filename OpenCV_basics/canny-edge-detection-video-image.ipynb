### Canny edge detection
- Used to find edges in a frame
- The Canny edge detector is an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images. It was developed by John F. Canny in 1986.
- The Canny edge detection algorithm is composed of 5 steps:
    1. Noise reduction;
    2. Gradient calculation;
    3. Non-maximum suppression;
    4. Double threshold;
    5. Edge Tracking by Hysteresis.
- Input
    - first argument is image
    - Second is minimum value.
    - Third argument is  maximum value. Third argument is aperture_size. It is the size of Sobel kernel used for find image gradients. By default it is 3. Last argument is L2gradient which specifies the equation for finding gradient magnitude.
- Sobel Kernel/Filter --> it creates an image emphasising edges


```python
# Edge detection image
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("standard_test_images\\standard_test_images\\1.tiff")
canny=cv2.Canny(img,90,90)
name=[img,canny]
titles=["img","canny"]
for i in range(0,len(name),1):
    plt.subplot(2,2,i+1)
    plt.imshow(name[i],cmap='gray'),plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
    cv2.imshow(titles[i],name[i])
    #if cv2.waitKey(0) & 0xFF==ord('q'):
     #   break
    
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
    
```

```python
# Edge detection live video stream and black-white color video live stream

import cv2
import numpy as np
capture=cv2.VideoCapture(0)
while True:
    ret,frame=capture.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    canny=cv2.Canny(frame,90,90)
    cv2.imshow("canny",canny)
    cv2.imshow("gray",gray)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
capture.release()
cv2.destroyAllWindows()
```
